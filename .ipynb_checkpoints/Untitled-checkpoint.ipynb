# Line Plot

```python
# importing matplotlib module 
from matplotlib import pyplot as plt 

# x-axis values 
x = [1, 2, 3, 4, 5] 

# Y-axis values 
y = [15, 8, 7, 3, 0] 

# Function to plot 
plt.plot(x, y) 

# function to show the plot 
plt.show() 
```

# Bar Plot

```python
# importing matplotlib module 
from matplotlib import pyplot as plt 

# x-axis values 
x = [5, 2, 9, 4, 7] 

# Y-axis values 
y = [10, 5, 8, 4, 2] 

# Function to plot 
plt.bar(x, y) 

# function to show the plot 
plt.show() 
```

# Histogram

```python
# importing matplotlib module 
from matplotlib import pyplot as plt 

# Y-axis values 
y = [10, 4,5, 8, 8, 1] 

# Function to plot histogram 
plt.hist(y) 

# Function to show the plot 
plt.show() 
```

# Scatter Plot

```python
# importing matplotlib module 
from matplotlib import pyplot as plt 

# x-axis values 
x = [5, 8, 9, 4, 8] 

# Y-axis values 
y = [10, 5, 8, 4, 2] 

# Function to plot scatter 
plt.scatter(x, y) 

# function to show the plot 
plt.show() 
```
